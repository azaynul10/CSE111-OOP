```python
#Task2
class Exam:
    def __init__(self,marks):
        self.marks = marks
        self.time = 60

    def examSyllabus(self):
        return "Maths , English"
    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"

class ScienceExam(Exam):
    def __init__(self,marks,time,*subjects):
        super().__init__(marks)
        self.time=time
        self.subject=''
        self.subjects=subjects
        self.no=2
        for i in self.subjects:
            self.no+=1
            self.subject+=i+', '
    def examSyllabus(self):
        return "Maths , English " + self.subject[:-2]
    def examParts(self):
        s=""

        for i in range(len(self.subjects)):
            s+="part "+str(i+2)+" - "+ self.subjects[i] +"\n"
        return "Part 1 - Maths\nPart 2 - English\n" + s
    def __str__(self):
        return "Marks: "+ str(self.marks) + " Time: "+ str(self.time)+" minutes Number of Parts: "+ str(self.no)


engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================')
architecture = ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())

```

```python
#Task3
class PokemonBasic:

    def __init__(self, name = 'Default', hp = 0, weakness = 'None', type = 'Unknown'):
        self.name = name
        self.hit_point = hp
        self.weakness = weakness
        self.type = type

    def get_type(self):
        return 'Main type: ' + self.type

    def get_move(self):
        return 'Basic move: ' + 'Quick Attack'

    def __str__(self):
        return "Name: " + self.name + ", HP: " + str(self.hit_point) + ", Weakness: " + self.weakness
class PokemonExtra(PokemonBasic):
    def __init__(self, name = 'Default', hp = 0, weakness = 'None', type = 'Unknown',secondary_type='Unknown',other_move=()):
        super().__init__(name,hp,weakness,type)
        self.secondary_type=secondary_type
        self.other_move=other_move


    def get_type(self):
        if self.secondary_type!="Unknown":
            return 'Main type: ' + self.type + ', Secondary type:' + self.secondary_type
        else:
            return 'Main type: ' + self.type

    def get_move(self):
        moves=""
        for i in self.other_move:
            moves+=i+', '

        return 'Basic move: ' + 'Quick Attack' + '\nOther move: ' + moves[:-2]

    def __str__(self):
        return "Name: " + self.name + ", HP: " + str(self.hit_point) + ", Weakness: " + self.weakness

print('\n------------Basic Info:--------------')
pk = PokemonBasic()
print(pk)
print(pk.get_type())
print(pk.get_move())

print('\n------------Pokemon 1 Info:-------------')
charmander = PokemonExtra('Charmander', 39, 'Water', 'Fire')
print(charmander)
print(charmander.get_type())
print(charmander.get_move())

print('\n------------Pokemon 2 Info:-------------')
charizard = PokemonExtra('Charizard', 78, 'Water', 'Fire', 'Flying', ('Fire Spin', 'Fire Blaze'))
print(charizard)
print(charizard.get_type())
print(charizard.get_move())
```

```python
#Task4
class Cakes:
    order_list={}
    feedback_dic={}
    def __init__(self,flavour,weight):
        self.flavour=flavour
        self.weight=weight
        self.sweet_lvl="Moderate "
        self.cream_type="Wipped cream"
        self.order=""
        self.price=(self.weight/1000)*1200
        self.order+=self.flavour+" Cake "+str(self.weight)+" gm"
        if self.order not in self.order_list:
            self.order_list[self.order]=1
        else:
            self.order_list[self.order]+=1
    def cake_details(self):
        print(f"Flavour: {self.flavour} cake, Weight: {self.weight} gm")
        print(f"Sweetness: {self.sweet_lvl} sugar, Cream type: {self.cream_type}")
        print("Price:",self.price,"taka")
    def add_customization(self,sweet_lvl,cream_type):
        self.sweet_lvl=sweet_lvl
        self.cream_type=cream_type
    @classmethod
    def give_feedbacks(cls,cake_name,feedback):
        cls.feedback_list=[]
        if feedback not in cls.feedback_list and cake_name not in cls.feedback_dic:
            cls.feedback_list.append(feedback)
            cls.feedback_dic[cake_name]=cls.feedback_list
        elif cake_name in cls.feedback_dic:
            cls.feedback_dic[cake_name].append(feedback)
        print("Thanks for your valuable feedback!")
    @classmethod
    def show_feedbacks(cls):
        print(cls.feedback_dic)
class Cheese_Cakes(Cakes):
    def __init__(self,flavour,weight,cake_type="Baked"):
        flavour+=" Cheese"
        super().__init__(flavour,weight)
        self.price=(self.weight/1000)*1500
        self.cream_type="Cream Cheese"
        self.order=""
        self.order+=self.flavour+" Cheese Cake "+str(self.weight)+" gm"

    def add_customization(self):
        print("Sorry! No customization available for cheese cakes.")
    @classmethod
    def give_feedbacks(cls,cake_name,feedback):
        super().give_feedbacks(cake_name,feedback)
        print("You will get 10% discount for your next purchase!")
    def cake_details(self):
        print(f"Flavour: {self.flavour} Cheese cake, Weight: {self.weight} gm")
        print(f"Sweetness: {self.sweet_lvl} sugar, Cream type: {self.cream_type}")
        print("Price:",self.price,"taka")
order_1=Cakes("Chocolate",500)
order_2=Cakes("Vanilla",800)
print("(1)**********************************")
order_1.cake_details()
print("(1.1)**********************************")
print(Cakes.order_list)
print("(2)**********************************")
order_2.add_customization("Zero","Butter Cream")
order_2.cake_details()
print("(3)**********************************")
Cakes.give_feedbacks("Chocolate Cake","Very Delicious")
Cakes.give_feedbacks("Chocolate Cake","Yummy")
print("(4)**********************************")
Cakes.show_feedbacks()
print("(5)**********************************")
ch_order1=Cheese_Cakes("Red velvet",700)
ch_order1.cake_details()
print("(6)**********************************")
ch_order1.add_customization()
print("(7)**********************************")
ch_order2=Cheese_Cakes("Blue Berry",900,"No Bake")
ch_order2.cake_details()
print("(8)**********************************")
print(Cakes.order_list)
print("(9)**********************************")
Cheese_Cakes.give_feedbacks("Red velvet Cheese Cake","Average")
Cheese_Cakes.give_feedbacks("Blue Berry Cheese Cake","Liked it")
print("(10)**********************************")
Cakes.show_feedbacks()
```
